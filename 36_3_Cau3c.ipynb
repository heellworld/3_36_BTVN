```python
from sklearn.feature_extraction.text import CountVectorizer
import os
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import nltk
nltk.download('punkt')
```

```python
# Define the directory path
directory_path = r"./Downloads/Data3/Data3/Data_Preprocessing/Data_Preprocessing.txt"
# Đọc nội dung từ tệp
with open(directory_path, 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
text
```

```python
# Tách thành danh sách các câu
sentences = nltk.sent_tokenize(text)
```

```python
sentences
```

```python
tr_idf_model  = TfidfVectorizer()
tf_idf_vector = tr_idf_model.fit_transform(sentences)
```

```python
tf_idf_array = tf_idf_vector.toarray()
tf_idf_array
```

```python
W = tr_idf_model.get_feature_names_out()
W
```

```python
df_tf_idf = pd.DataFrame(tf_idf_array, columns = W)
df_tf_idf
```
